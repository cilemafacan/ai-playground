```python
import PIL
import requests
import torch
from io import BytesIO

from diffusers import StableDiffusionInstructPix2PixPipeline


image = PIL.Image.open("yaprak.jpeg")
image.thumbnail((768,768))
pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(
    "timbrooks/instruct-pix2pix", torch_dtype=torch.float16
)
pipe = pipe.to("cuda")

prompt = "make the background black, red leaves"
negative_prompt = " black leaves"
image = pipe(prompt=prompt, image=image, num_inference_steps=50, negative_prompt=negative_prompt).images[0]
```
